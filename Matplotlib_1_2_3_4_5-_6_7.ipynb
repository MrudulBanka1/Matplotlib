```python
blood_sugar =[121,90,100, 150, 129, 80, 102, 120, 220, 96]
blood_sugarw =[129,94,110, 50, 120, 89, 126, 180, 200, 80]
```

```python
plt.hist(blood_sugar)
```

```python
plt.hist(blood_sugar, bins =3, rwidth =0.95)
```

```python
plt.hist(blood_sugar, bins =[80,100,125,250],rwidth=0.95)
```

```python
plt.hist(blood_sugar, bins =[80,100,125,250],rwidth=0.95, color='g', histtype='step')
```

```python
plt.hist([blood_sugar, blood_sugarw], bins =[80,100,125,250],rwidth=0.95, color=['r','g'], label=['men','women'])
plt.legend()
plt.ylabel('No fo patients')
plt.xlabel('sugar level')
plt.title('Comparison of blood sugar levels')
```

```python
plt.hist([blood_sugar, blood_sugarw], bins =[80,100,125,250],rwidth=0.95, color=['r','g'], label=['men','women'], orientation='horizontal')
plt.legend()
plt.xlabel('No fo patients')
plt.ylabel('sugar level')
plt.title('Comparison of blood sugar levels')
```

```python
values =[100,200,450,1000]
exp=['Fun','Fuel','Food','Flat']
```

```python
plt.pie(values, labels=exp)
plt.axis('equal')
plt.show()
```

```python
plt.pie(values, labels=exp, radius=1.5, autopct='%0.2f%%', shadow=True)
plt.axis('equal')
plt.show()
```

```python
plt.pie(values, labels=exp, radius=1.5, autopct='%0.2f%%', shadow=True, explode=[0,1,0,0], startangle=45)
plt.axis('equal')
#plt.show()
plt.savefig("Piechart.png", bbox_inches='tight', pad_inches=2,transparent=True)
```
